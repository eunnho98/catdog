```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
```

```python
model = torchvision.models.resnet50(pretrained=True)
model
```

```python
num_ftrs = model.fc.in_features # 모델의 fully connected 레이어의 입력 특성 수를 가져옴

model.fc = nn.Sequential( # 기존 모델의 fully connected 레이어를 새로운 레이어로 대체
    nn.Dropout(0.5),
    nn.Linear(num_ftrs, 1024),
    nn.Dropout(0.2),
    nn.Linear(1024, 512),
    nn.Dropout(0.1),
    nn.Linear(512, 1),
    nn.Sigmoid()
)
```
